```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import seaborn as sns

#Media Narrative is that LeBron James and Kevin Duran have dominated the NBA for the best part of 15 years - Let's see who comes out on top.
```

```python
Player_stats= pd.read_csv("all_seasons.csv")
```

```python
Player_stats.rename(columns={"player_name" : "Player Name", "pts" : "Points", "reb" : "Rebounds", "ast" : "Assists"}, inplace=True)
good = []
Player_stats
```

```python
players = Player_stats[["Player Name", "Points", "Rebounds", "Assists", "season"]]
averages = players.groupby("Player Name").mean()
best_averages = averages.sort_values(["Points","Rebounds", "Assists"], ascending = False)
```

```python
LeBron = players.loc[players["Player Name"] == "LeBron James"]
```

```python
Durant = players.loc[players["Player Name"] == "Kevin Durant"]
```

```python
LeBron
```

```python
Durant
```

```python
compare = pd.concat([LeBron,Durant])
```

```python
compare
```

```python
#Group and use the mean function to return the career averages
average= compare.groupby('Player Name').mean()
```

```python
average
```

```python
compare = average.swapaxes("index", "columns")
```

```python
plt.rcParams.update({'axes.facecolor' : 'yellow'})
compare.plot.bar()
plt.title("LBJ Vs KD", fontsize=25, color = "Black")
plt.legend(facecolor ="white", edgecolor= "blue", fontsize=15)
plt.xticks(fontsize=20, color= "black", rotation = "horizontal")
plt.yticks(fontsize=20, color ="black")

plt.show()


```

The data analysis here shows that LeBron James has the higher individual stats when it comes to Points, Rebounds and especially assists when compared to Kevin Durant.
